```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(path + "/books.csv")
books.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), annot=False, ax=ax)
plt.xlabel("Column")
plt.ylabel("Missing values")
plt.show()
```

```python
import numpy as np
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]
correlation_matrix = books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
# Create an explicit copy
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25].copy()
```

```python
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = np.where(
    book_missing_25_words["subtitle"].isna(),
    book_missing_25_words["title"],
    book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(
    " ".join, axis=1)
```

```python
(
    book_missing_25_words
    .drop(columns=["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("book_cleaned.csv", index = False)
)
```
